## Importing / exporting data

```python
## Import csv file
# Import packages
import os
import json

# Declare function
def import_csv(folder_name, file_name):
    """Imports csv file as DataFrame"""
    csv_dataframe = pd.read_csv(os.path.join(os.getcwd(), folder_name, file_name), sep=',')
    return csv_dataframe

# Run function
import_csv('data', 'data.csv')
```

```python
## Import json file
# Import packages
import os
import json

# Declare function
def import_json(folder_name, file_name):
    """Imports json file as DataFrame"""
    json_data = json.load(open(os.path.join(os.getcwd(), folder_name, file_name)))
    json_dataframe = pd.DataFrame.from_dict(json_normalize(json_data), orient='columns')
    return json_dataframe

# Run function
import_json('data', 'data.json')
```

```python
## Save dataframe as csv
# Import packages
import os
import pandas as pd
import numpy as np
from sklearn.datasets import load_diabetes

# Load data
diabetes = load_diabetes()
data = pd.DataFrame(data= np.c_[diabetes['data'], \
                                     diabetes['target']],
                        columns= ['age','sex','bmi','bp',
                                  's1','s2','s3',
                                  's4','s5','s6',
                                  'target'] 
                   )

# Declare function
def save_dataframe_as_csv(dataframe, file_name):
    """Save dataframe as csv in output folder within current working directory"""
    if not os.path.exists('output/'):
        os.makedirs('output')
    dataframe.to_csv(os.path.join('output', file_name), sep=',', index=False)
    
# Run function
save_dataframe_as_csv(data, 'output.csv')
```

```python
## Normalize and converts json to dataframe
# Import packages
import pandas as pd
from pandas.io.json import json_normalize

# Load data
# data = [
#   {
#     'name': {
#       'first': 'vikash',
#       'last': 'singh'
#     },
#     'age': 29
#   },
#   {
#     'name': {
#       'first': 'satyam',
#       'last': 'singh'
#     },
#     'age': 14
#   }
# ]

data = [{'state': 'Florida',
          'shortname': 'FL',
          'info': {
               'governor': 'Rick Scott'
          },
          'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
         {'state': 'Ohio',
          'shortname': 'OH',
          'info': {
               'governor': 'John Kasich'
          },
          'counties': [{'name': 'Summit', 'population': 1234},
                       {'name': 'Cuyahoga', 'population': 1337}]}]

# Declare function
def convert_json_to_df(json_data):
    """Normalize and converts json to Pandas DataFrame"""
    dataframe = json_normalize(json_data
                               ,record_path='counties'
                               ,meta=['state', 'shortname',['info','governor']]
                               ,record_prefix='record.'
                               ,meta_prefix='meta.'
                              )
# record_path: path to a list of records
# meta: fields to use as metadata for each record in resulting table
    return dataframe

# Run function
convert_json_to_df(data)
```

## Converting datafiles

```python
## Convert json to dataframe
# Import packages
import pandas as pd
from pandas.io.json import json_normalize

# Load data
data = [
  {
    'name': {
      'first': 'vikash',
      'last': 'singh'
    },
    'age': 29
  },
  {
    'name': {
      'first': 'satyam',
      'last': 'singh'
    },
    'age': 14
  }
]

# Declare function
def convert_json_to_df(json_data):
    """Converts json to Pandas DataFrame"""
    dataframe = pd.DataFrame.from_dict(json_normalize(json_data), orient='columns')
    return dataframe

# Run function
convert_json_to_df(data)
```

## Visualisation

```python
## Plotting histogram
# Import packages
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes

# Load data
diabetes = load_diabetes()
data = pd.DataFrame(data= np.c_[diabetes['data'], \
                                     diabetes['target']],
                        columns= ['age','sex','bmi','bp',
                                  's1','s2','s3',
                                  's4','s5','s6',
                                  'target'] 
                   )

# Declare function
def plot_histogram(dataframe, column_name):
    """Plot histogram"""
    dataframe[column_name].plot('hist', bins=10, logx=False, logy=False)
    plt.show()

# Run function
plot_histogram(data,'age')
```

```python
## Plotting scatter plot
# Import packages
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes

# Load data
diabetes = load_diabetes()
data = pd.DataFrame(data= np.c_[diabetes['data'], \
                                     diabetes['target']],
                        columns= ['age','sex','bmi','bp',
                                  's1','s2','s3',
                                  's4','s5','s6',
                                  'target'] 
                   )

# Declare function
def plot_scatter_plot(dataframe, column_name_1, column_name_2):
    """Plot scatter plot"""
    dataframe.plot(kind='scatter', x=column_name_1, y=column_name_2)
    plt.show()
    
# Run function
plot_scatter_plot(data,'bmi', 'bp')
```

```python
## Plotting box plot
# Import packages
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes

# Load data
diabetes = load_diabetes()
data = pd.DataFrame(data= np.c_[diabetes['data'], \
                                     diabetes['target']],
                        columns= ['age','sex','bmi','bp',
                                  's1','s2','s3',
                                  's4','s5','s6',
                                  'target'] 
                   )

# Declare function
def plot_box_plot(dataframe, list_of_column_names):
    """Plot box plot"""
    dataframe[list_of_column_names].plot(kind='box',subplots=False)
    plt.show()
    
# Run function
plot_box_plot(data,['bmi','age','s1'])
```

```python
## Aggregating data via groupby
# Import packages
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer

# Load data
breast_cancer = load_breast_cancer()
data = pd.DataFrame(data= np.c_[breast_cancer['data'], \
                                 breast_cancer['target']],
                    columns= ['mean radius', 'mean texture', 'mean perimeter', 'mean area',
                           'mean smoothness', 'mean compactness', 'mean concavity',
                           'mean concave points', 'mean symmetry', 'mean fractal dimension',
                           'radius error', 'texture error', 'perimeter error', 'area error',
                           'smoothness error', 'compactness error', 'concavity error',
                           'concave points error', 'symmetry error', 'fractal dimension error',
                           'worst radius', 'worst texture', 'worst perimeter', 'worst area',
                           'worst smoothness', 'worst compactness', 'worst concavity',
                           'worst concave points', 'worst symmetry', 'worst fractal dimension', 
                            'target'] 
                    )

# Declare function
def group_by_data(dataframe, groupby_column, list_of_columns, list_of_functions):
    """Groupby data by using an aggregate function"""
    groupby_dataframe = dataframe.groupby(groupby_column)[list_of_columns].agg(list_of_functions)
    return groupby_dataframe
    
# Run function
group_by_data(data, 'target', ['mean radius', 'mean texture'], ['median','mean','count'])
```
